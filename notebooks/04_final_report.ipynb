```python
# Cell 2: Setup & Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure Plotting Style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Market Analysis (The "Why")
Our initial analysis (EDA) revealed distinct behaviors for each asset:
*   **TSLA:** High Volatility, High Reward. The "Growth Engine."
*   **BND:** Low Volatility, Stable Returns. The "Safety Net."
*   **SPY:** Moderate Volatility. The "Market Benchmark."

The chart below illustrates the cumulative return of $1 invested in each asset over the analysis period.

# Cell 4: Cumulative Returns Plot
# Load Data
df = pd.read_csv('../data/processed/cleaned_close_prices.csv', index_col=0, parse_dates=True)

# Calculate Cumulative Returns (Growth of $1)
# We add 1 to the percentage change, then multiply them all together (cumprod)
cumulative_returns = (1 + df.pct_change()).cumprod()

# Plot
cumulative_returns.plot()
plt.title('Growth of $1 Investment (2015-2024)')
plt.ylabel('Investment Value ($)')
plt.show()

## 2. Forecasting (The "Outlook")
Our ARIMA models indicate that Tesla's volatility is expected to continue. While long-term growth is possible, the short-term price action is highly unpredictable (MAPE ~28%).

**Implication:** Relying solely on Tesla is too risky. We must hedge this risk with BND and SPY to protect capital.

## 3. Portfolio Optimization (The "Solution")
Using Modern Portfolio Theory, we simulated 10,000 portfolio combinations to find the mathematical optimum.

### The Optimal Portfolio (Max Sharpe Ratio)
This portfolio offers the best return for every unit of risk taken.

```python
# Cell 7: Recommended Allocation Visualization


optimal_weights = {
    'TSLA': 0.103,  # <--- UPDATE ME
    'BND': 0.59,   # <--- UPDATE ME
    'SPY': 0.29    # <--- UPDATE ME
}


# Create a Pie Chart
plt.figure(figsize=(7, 7))
plt.pie(optimal_weights.values(), 
        labels=optimal_weights.keys(), 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=['red', 'green', 'orange'])
plt.title('Recommended Portfolio Allocation (Max Sharpe)')
plt.show()
```
